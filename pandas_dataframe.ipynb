```python
# .astype()
import pandas as pd
names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation',\
         'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week',\
         'native-country', 'salary']
data_c = pd.read_csv('adult.csv', header = None, names = names)
# Information regarding dtaset
data_c.info()
```

```python
# In normal scenario pandas decide the suitable datatype for each column but sometimes
# it might affect the memory.
# for example : Most of the time gender is having two category that is male and female
# and sometimes pandas decide to give gender calumn as object as datatype which consumes
# higher memory as compare to category datatype
data_c['sex'] = data_c['sex'].astype('category')
data_c.info()   # Memory change to 3.7 to 3.5
# Same goes for merital status
data_c['marital-status'] = data_c['marital-status'].astype('category')
data_c.info()   # Memory change to 3.5 to 3.3
```
